```python
import pandas as pd
```

```python
df = pd.read_csv("ready_dataset_fromStep3.csv")
```

<hr>

# Selection

* 3 age groups: 25-34, 35-44, 45-54

* 4 income classes: lower-middle, middle, upper-middle, upper

* 4 ethnicities: white, asian, hispanic, black

<hr>

# Age groups

```python
df = df[
    (df['applicant_age'] == '25-34') | (df['applicant_age'] == '35-44') | (df['applicant_age'] == '45-54')
]
```

```python
selected_age_order = [
    '25-34', '35-44', '45-54'
]

df['applicant_age'] = pd.Categorical(
    df['applicant_age'],
    categories=selected_age_order,
    ordered=True
)
```

```python
df['applicant_age'].value_counts()
```

<hr>

# Income groups

```python
df = df[
    (df['income_class'] == 'Lower-middle') | (df['income_class'] == 'Middle') | (df['income_class'] == 'Upper-middle') | (df['income_class'] == 'Upper')
]
```

```python
selected_income_order = [
    'Lower-middle', 'Middle', 'Upper-middle', 'Upper'
]

df['income_class'] = pd.Categorical(
    df['income_class'],
    categories=selected_income_order,
    ordered=True
)
```

```python
df['income_class'].value_counts()
```

<hr>

# Ethnicity groups

```python
df = df[
    (df['ethnicity_race'] == 'Asian') | (df['ethnicity_race'] == 'White') | (df['ethnicity_race'] == 'Hispanic or Latino') | (df['ethnicity_race'] == 'Black or African American')
]
```

```python
df['ethnicity_race'].value_counts()
```
