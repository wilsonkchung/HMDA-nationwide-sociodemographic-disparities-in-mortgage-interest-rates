```python
import pandas as pd

import statsmodels.formula.api as smf
```

```python
df = pd.read_csv("df_trimmed.csv")
```

```python
df.shape
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df.head()
```

```python
model = smf.ols(
    formula="""
    interest_rate ~
    C(applicant_race_1)
    + C(applicant_ethnicity_1)
    + C(applicant_sex)
    + log_loan_amount
    + log_income
    + debt_to_income_ratio
    + loan_to_value_ratio
    + loan_term
    + C(loan_type)
    + C(loan_purpose)
    + C(state_code)
    + C(year)
    """,
    data=df
).fit(cov_type="HC3")  # robust SEs (important)
```
